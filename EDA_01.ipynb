Analysing Sentence :  
target_id - (string) An arbitrary identifier.  
In train_sequences.csv, this is formatted as pdb_id_chain_id, where pdb_id is the id of the entry in the Protein Data Bank and chain_id is the chain id of the monomer in the pdb file.

```python
# Split the 'target_id' column into 'pdb_id' and 'chain_id'
train_sequences[['pdb_id', 'chain_id']] = train_sequences['target_id'].str.split('_', expand=True)

pdb_id_unique = train_sequences['pdb_id'].nunique()
chain_id_unique = train_sequences['chain_id'].nunique()

pdb_id_total = train_sequences['pdb_id'].count()
chain_id_total = train_sequences['chain_id'].count()

print(f"There are {pdb_id_unique} unique pdb_id's in the dataset out of total {pdb_id_total} pdb_id's.")
print(f"There are {chain_id_unique} unique chain_id's in the dataset out of total {chain_id_total} chain_id's.")
```

```python
# Process for pdb_id
repeated_pdb_ids = (
    train_sequences
    .groupby('pdb_id')
    .size()
    .reset_index(name='total')
    .query('total > 1')
    .sort_values(by='total', ascending=False)
)
top_repeated_pdb_ids = repeated_pdb_ids.head(10)

print("In total, there are", repeated_pdb_ids.shape[0], "pdb_id's that are repeated in train_sequences.")
print("Top 10 most repeated pdb_id's:")
display(top_repeated_pdb_ids)

# Process for chain_id
repeated_chain_ids = (
    train_sequences
    .groupby('chain_id')
    .size()
    .reset_index(name='total')
    .query('total > 1')
    .sort_values(by='total', ascending=False)
)
top_repeated_chain_ids = repeated_chain_ids.head(10)

print("In total, there are", repeated_chain_ids.shape[0], "chain_id's that are repeated in train_sequences.")
print("Top 10 most repeated chain_id's:")
display(top_repeated_chain_ids)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train_seq = pd.read_csv('train_sequences.csv')

# Calculate sequence lengths
train_seq['sequence_length'] = train_seq['sequence'].apply(len)

# Print statistics
print("Sequence Length Statistics:")
print(train_seq['sequence_length'].describe())

# Plot distribution
bin_edges = np.linspace(train_seq['sequence_length'].min(), train_seq['sequence_length'].max(), 20)
train_seq['length_bin'] = pd.cut(train_seq['sequence_length'], bins=bin_edges)
bin_counts = train_seq['length_bin'].value_counts().sort_index()
bin_labels = [f"{int(interval.left)}-{int(interval.right)}" for interval in bin_counts.index]

plt.figure(figsize=(12, 6))
sns.barplot(x=bin_labels, y=bin_counts.values, color='blue')
plt.title('Sequence Length Distribution', fontsize=14)
plt.xlabel('Sequence Length Bins', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load train_sequences.csv
train_seq = pd.read_csv('train_sequences.csv')

# Calculate sequence lengths
train_seq['sequence_length'] = train_seq['sequence'].apply(len)

# Print statistics
print("Sequence Length Statistics:")
print(train_seq['sequence_length'].describe())

# Define the histogram plotting function
def plot_length_histogram(df, desc, ax=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 5))
    sns.histplot(df['sequence_length'], kde=True, ax=ax, color='blue')
    ax.set_title(f'{desc} Sequence Length Frequency', fontsize=14)
    ax.set_xlabel('Sequence Length', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    if ax is None:  # Save only if a new figure was created
        plt.savefig(f'Figures/{desc}_plot_length_histogram.png', dpi=300, bbox_inches='tight')
        plt.show()

# Plot histogram for train sequences
plot_length_histogram(train_seq, 'Train')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D plotting

# Load the data
train_labels = pd.read_csv("train_labels.csv")
train_sequences = pd.read_csv("train_sequences.csv")

# Define the target_id you want to plot
target_id = "1SCL_A"  # <-- You can change this to any valid target_id

# Filter train_labels by target_id
filtered_labels = train_labels[train_labels['ID'].str.startswith(target_id)]

# Sort by residue id to maintain sequence order
filtered_labels = filtered_labels.sort_values(by='resid')

# Extract coordinates
x = filtered_labels['x_1']
y = filtered_labels['y_1']
z = filtered_labels['z_1']
resid = filtered_labels['resid']
resname = filtered_labels['resname']

# Plot in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for points
sc = ax.scatter(x, y, z, c=resid, cmap='viridis', s=50, edgecolor='k')

# Connect points with a line
ax.plot(x, y, z, color='blue', linewidth=1.5, label='Residue Path')

# Annotate each point with the resname and resid
for i, (xi, yi, zi, rn, ri) in enumerate(zip(x, y, z, resname, resid)):
    ax.text(xi, yi, zi, f'{rn}{ri}', fontsize=8)

# Labels and title
ax.set_xlabel('x_1')
ax.set_ylabel('y_1')
ax.set_zlabel('z_1')
ax.set_title(f"3D Coordinates of Residues for Target: {target_id}")
plt.colorbar(sc, label='Residue ID')
ax.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
train_labels = pd.read_csv("train_labels.csv")
train_sequences = pd.read_csv("train_sequences.csv")

# Get first 10 unique target_ids from train_sequences
target_ids = train_sequences['target_id'].unique()[:10]

# Setup subplots: 5 rows x 2 cols
fig = plt.figure(figsize=(16, 20))

for idx, target_id in enumerate(target_ids):
    # Subplot position
    ax = fig.add_subplot(5, 2, idx + 1, projection='3d')
    
    # Filter and sort
    filtered_labels = train_labels[train_labels['ID'].str.startswith(target_id)]
    filtered_labels = filtered_labels.sort_values(by='resid')
    
    # Extract data
    x = filtered_labels['x_1']
    y = filtered_labels['y_1']
    z = filtered_labels['z_1']
    resid = filtered_labels['resid']
    resname = filtered_labels['resname']
    
    # Plot scatter and line
    sc = ax.scatter(x, y, z, c=resid, cmap='viridis', s=40, edgecolor='k')
    ax.plot(x, y, z, color='blue', linewidth=1.0)
    
    # Annotate residues (optional: can be commented out if cluttered)
    for xi, yi, zi, rn, ri in zip(x, y, z, resname, resid):
        ax.text(xi, yi, zi, f'{rn}{ri}', fontsize=6)
    
    # Titles and labels
    ax.set_title(f"{target_id}", fontsize=10)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

# Adjust layout
plt.tight_layout()
plt.show()
```
