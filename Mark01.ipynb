```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_sequences=pd.read_csv('train_sequences.csv')
train_labels=pd.read_csv('train_labels.csv')
val_sequences=pd.read_csv('validation_sequences.csv')
val_labels=pd.read_csv('validation_labels.csv')
```

```python
plt.figure(figsize=(12,5))
train_sequences['length'] = train_sequences['sequence'].str.len()
sns.boxplot(x='length', data=train_sequences)
plt.xlabel("Sequence length")
plt.ylabel("Sequence")
plt.title("Boxplot of sequence length values (with outliers)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,5))
train_sequences['length'] = train_sequences['sequence'].str.len()
sns.boxplot(x='length', data=train_sequences, showfliers=False)
plt.xlabel("Sequence length")
plt.ylabel("Sequence")
plt.title("Boxplot of sequence length values (without outliers)")
plt.grid(True)
plt.show()
```
