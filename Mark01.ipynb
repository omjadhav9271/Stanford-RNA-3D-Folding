```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group and count occurrences of each resname
resname_counts = train_labels.groupby('resname').size().reset_index(name='total')

plt.figure(figsize=(12, 5))
# Assign 'resname' to both x and hue for consistent coloring
ax = sns.barplot(x='resname', y='total', data=resname_counts, hue='resname', palette='Blues', dodge=False)

# Annotate each bar with its count
for idx, row in resname_counts.iterrows():
    ax.text(idx, row['total'], row['total'], ha='center', va='bottom')

plt.xlabel("Resname")
plt.ylabel("Count")
plt.title("Barplot of resname occurrences in train_labels")

# Remove the legend if it exists to avoid redundancy
legend = ax.get_legend()
if legend is not None:
    legend.remove()

sns.set_theme(style="whitegrid")
plt.show()
```

Analysing Sentence :  
target_id - (string) An arbitrary identifier.  
In train_sequences.csv, this is formatted as pdb_id_chain_id, where pdb_id is the id of the entry in the Protein Data Bank and chain_id is the chain id of the monomer in the pdb file.

```python
# Split the 'target_id' column into 'pdb_id' and 'chain_id'
train_sequences[['pdb_id', 'chain_id']] = train_sequences['target_id'].str.split('_', expand=True)

pdb_id_unique = train_sequences['pdb_id'].nunique()
chain_id_unique = train_sequences['chain_id'].nunique()

pdb_id_total = train_sequences['pdb_id'].count()
chain_id_total = train_sequences['chain_id'].count()

print(f"There are {pdb_id_unique} unique pdb_id's in the dataset out of total {pdb_id_total} pdb_id's.")
print(f"There are {chain_id_unique} unique chain_id's in the dataset out of total {chain_id_total} chain_id's.")
```

```python
# Process for pdb_id
repeated_pdb_ids = (
    train_sequences
    .groupby('pdb_id')
    .size()
    .reset_index(name='total')
    .query('total > 1')
    .sort_values(by='total', ascending=False)
)
top_repeated_pdb_ids = repeated_pdb_ids.head(10)

print("In total, there are", repeated_pdb_ids.shape[0], "pdb_id's that are repeated in train_sequences.")
print("Top 10 most repeated pdb_id's:")
display(top_repeated_pdb_ids)

# Process for chain_id
repeated_chain_ids = (
    train_sequences
    .groupby('chain_id')
    .size()
    .reset_index(name='total')
    .query('total > 1')
    .sort_values(by='total', ascending=False)
)
top_repeated_chain_ids = repeated_chain_ids.head(10)

print("In total, there are", repeated_chain_ids.shape[0], "chain_id's that are repeated in train_sequences.")
print("Top 10 most repeated chain_id's:")
display(top_repeated_chain_ids)
```
