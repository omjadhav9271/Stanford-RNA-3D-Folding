```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=train_sequences['sequence_length'])
plt.title('Boxplot of Sequence Lengths in Train Sequences')
plt.xlabel('Sequence Length')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=train_sequences['sequence_length'], showfliers=False)
plt.title('Boxplot of Sequence Lengths in Train Sequences (Without Outliers)')
plt.xlabel('Sequence Length')
plt.show()
```
